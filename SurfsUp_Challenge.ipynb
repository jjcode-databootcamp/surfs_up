```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
style.use('fivethirtyeight')
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==6).all()
```

```python
#  3. Convert the June temperatures to a list.
june_temp_list = session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==6).all()

#return the list in juypter
june_temp_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_temp_list, columns=['Date','June Temps'])

#return the dataframe in juypteR
june_temp_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

```python
#sort the June data by date from 2010 to 2017
june_temp_sorted_df = june_temp_df.sort_values(by='Date', ascending=True)

#set the index to date so the chart x axsis will be the date 
june_temp_sorted_df.set_index(june_temp_sorted_df['Date'], inplace=True)

june_temp_sorted_df
```

```python
#June Weather from year 2010 to 2017 
june_temp_sorted_df.plot(figsize=(18, 8))
plt.gcf().subplots_adjust(bottom=0.15)

# Add graph Title
plt.title('June Temperature Range Over Time', fontsize=18)

#Save graph 
plt.savefig('Resources/June_Temp_Chart.png');
```

```python
# June Temp by Frequency 
june_temp_df.plot.hist(figsize=(9, 8))
plt.gcf().subplots_adjust(bottom=0.15)

# Add graph Title
plt.title('June Temperature Frequency', fontsize=18)

#Save graph 
plt.savefig('Resources/June_Temp_Hist.png');

```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
# 7. Convert the December temperatures to a list.
# steps 6 and 7 can be accomplished in 2 step as you can see from 3 and 4 from above 
dec_temp_list = session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==12).all()

#return the June temp in juypter 
dec_temp_list
```

```python
#Dec Weather from year 2010 to 2017 
dec_temp_sorted_df.plot(figsize=(18, 8))
plt.gcf().subplots_adjust(bottom=0.15)

# Add graph Title
plt.title('December Temperature Range Over Time', fontsize=18)

#Save graph 
plt.savefig('Resources/Dec_Temp_Chart.png');
```
